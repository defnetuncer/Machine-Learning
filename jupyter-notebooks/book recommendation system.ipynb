### INTRODUCTION

A recommender system or a recommendation system is a subclass of information filtering system that seeks to predict the "rating" or "preference" a user would give to an item. In today's world it is essential for businesses to make personalized recommendations. ### COLLABORATIVE FILTERING
The Collaborative Filtering (CF) algorithms are based on the idea that if two clients have similar rating history then they will behave similarly in the future. [1] One of the preferred approaches is to use k-Nearest-Neighborhood (kNN) classifier, which is dependent on defining an appropriate similarity or distance measure. [2]

### SIMILARITY MEASURES
In this section various similarity measures have been explained. [3][4]
### Cosine Similarity (User-Based CF)
Cosine similarity finds how two vectors are related to each other using measuring cosine angle between these vectors. The major drawback with cosine similarity is that it considers null preferences as negative preference.

\begin{equation*}
CosSim(u,v) = \frac{\sum_{i∈I}(r_{u,i})(r_{v,i})}{\sqrt{\sum_{i∈I}{(r_{u,i})}^2}\sqrt{\sum_{i∈I}{{(r_{v,i})}^2}}}
\end{equation*}

### Correlation Similarity (User-Based CF)
This method is used to find linear correlation between two vectors. Pearson Correlarion Coefficient results a value between -1 and +1. -1 represents a negative co relation while +1 represents high positive correlation. 0 value shows no relation sometimes called zero order correlation. Where $\overline{r}_{u,I}$ and $\overline{r}_{v,I}$ represents average rating of user u and user v, respectively, for co-rated items represented by set I.

\begin{equation*}
CorrelationSim(u,v) = \frac{\sum_{i∈I}(r_{u,i}-\overline{r}_{u,I})(r_{v,i}-\overline{r}_{v,I})}{\sqrt{\sum_{i∈I}{(r_{u,i}-\overline{r}_{u,I})}^2}\sqrt{\sum_{i∈I}{{(r_{v,i}-\overline{r}_{v,I})}^2}}}
\end{equation*}

### Adjusted Cosine Similarity (User-Based CF)
Adjusted cosine similarity is slightly different from Pearson Correlation; Pearson Correlation considers the average rating of user u for co-rated the items. Adjusted cosine similarity subtracts the average rating of user u for all the items rated by user u. Where $\overline{r_i}$ average rating of user i for all items rated by user i itself.

\begin{equation*}
AdjustedCosSim(u,v) = \frac{\sum_{i∈I}(r_{u,i}-\overline{r}_u)(r_{v,i}-\overline{r}_v)}{\sqrt{\sum_{i∈I}{(r_{u,i}-\overline{r}_u)}^2}\sqrt{\sum_{i∈I}{{(r_{v,i}-\overline{r}_v)}^2}}}
\end{equation*}

### CONCLUSION
In this paper a Book Recommendation System is built using Book-Crossing Dataset. Distance Weighted k-Nearest Neighbor and k-Nearest Neighbor Classifiers are implemented based on User-Based Collaborative Filtering while comparing three different similarity measures. Collaborative filtering methods suffer from issues like cold start, scalability and sparsity and User-Based CF requires large amount of user data. Since the 60 percent of the users rated only one book, collaborative filtering based on user would be a bad choice. Also while testing, implicit data should be discarded since it's not a rating. These conditions effect negatively on Mean Absolute Error. k-Nearest Neighbors Algorithm is a good choice for Recommendation Systems but execution time takes too long. Overall in this paper, steps to follow for a Machine Learning research is studied, data analysed and pros-cons of classifiers are computed and determined.
